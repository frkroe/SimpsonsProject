## Paso 1: Plotear las palabras más usadas en tiempo-real

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from time import sleep

while True: 
## leer y sortear datos por cuenta y guardar los 10 de valores más grandes
    df = pd.read_csv(r"C:\Users\franz\Documents\GitHub\SimpsonsProject\Project3Bart\Python\app\cuentas.csv", delimiter=",")
    datos = df.sort_values(by=['Cuenta'],ascending=False).nlargest(10,'Cuenta')
## plotear datos
    plot = datos.plot.bar(x='Palabra', y='Cuenta')       
    plt.xlabel("Cuenta")
    plt.ylabel("Palabra")
    plt.title('las palabras más habituales en Simpsons')
##  borrar el output del gráfico y 30 segundos para seguir ejecutando el bucle while (refrescar los datos del gráfico)
    clear_output(wait=True)
    plt.show()
    sleep(30)
```

## Paso 2: Plotear las ímagenes

```python
from shutil import copytree
import os
from IPython.display import Image, display, clear_output
## plotear imagenes con su nombre de carpeta
dictImg = {}
for (root, dirs, files) in os.walk('../Python/app'):
    for file in files:
        if file.endswith(".png"):
            im = os.path.join(root, file)
            dictImg[file]= im
for nombre, img in dictImg.items():
    print(nombre.replace(".png", ""))
    display(Image(img))
    clear_output(wait=True)




```
